```python
import pandas as pd
import os
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
caseType = 'Death'
#caseType = 'Discharge'
seperator = '_'
deathCaseUHID = ["RNEH.0000008375", "RNEH.0000011301", "RNEH.0000012581", "RNEH.0000013713", "RSHI.0000012088", "RSHI.0000013287", "RSHI.0000014720"
, "RSHI.0000015178", "RSHI.0000015211", "RSHI.0000015691", "RSHI.0000016373", "RSHI.0000017471", "RSHI.0000017472", "RSHI.0000019707"
, "RSHI.0000021953", "RSHI.0000023451"]
#deathCaseUHID = ["RSHI.0000021953"]
```

```python
data = pd.DataFrame()
plt.rcParams["figure.figsize"] = (20,10)
for uhid in deathCaseUHID:
    fileName = path+"/Death_Cases/"+uhid+"/"+caseType+seperator+uhid+seperator+'intermediate_checkpoint_new_5.csv'
    print (fileName)
    data = pd.read_csv(fileName)
    fig, (ax1, ax2,ax3,ax4,ax5,ax6) = plt.subplots(6, sharex=True)
    fig.suptitle('Distributions for '+uhid)
    ax1.set(xlabel='Minutes', ylabel='Continuous Signals')
    ax2.set(xlabel='Minutes', ylabel='Intermittent Intake')
    ax3.set(xlabel='Minutes', ylabel='Intermittent Output')
    ax4.set(xlabel='Minutes', ylabel='Intermittent Vitals')
    ax5.set(xlabel='Minutes', ylabel='Intermittent Ventilator')
    ax6.set(xlabel='Minutes', ylabel='Intermittent Anthropometry')   
    
    x = data['Unnamed: 0']
    y = data.spo2
    ax1.plot( x, 'spo2',  data=data, marker='', color='skyblue', linewidth=2)
    ax1.plot( x, 'pulserate',  data=data, marker='', color='red', linewidth=2)
    ax1.plot( x, 'heartrate',  data=data, marker='', color='red', linewidth=2)

    ax2.plot(  x, 'total_intake', data=data, marker='', color='red', linewidth=2)
    ax2.plot(  x, 'totalparenteralvolume', data=data, marker='', color='indigo', linewidth=2)
    ax2.plot(  x, 'tpn-tfl', data=data, marker='', color='chocolate', linewidth=2)

    ax3.plot(  x, 'stool_day_total', data=data, marker='', color='deepskyblue', linewidth=2)
    ax3.plot(  x, 'stool_passed', data=data, marker='', color='pink', linewidth=2)
    ax3.plot(  x, 'urine', data=data, marker='', color='slategray', linewidth=2)
    ax3.plot(  x, 'urine_per_hour', data=data, marker='', color='lime', linewidth=2)
    
    
    ax4.plot(  x, 'mean_bp', data=data, marker='', color='green', linewidth=2)
    ax4.plot(  x, 'rbs', data=data, marker='', color='red', linewidth=2)
    ax4.plot(  x, 'new_ph', data=data, marker='', color='blue', linewidth=2)

    ax5.plot( x, 'fio2', data=data, marker='', color='darkgoldenrod', linewidth=2)
    ax5.plot( x, 'peep', data=data, marker='', color='royalblue', linewidth=2)
    ax5.plot( x, 'pip', data=data, marker='', color='olive', linewidth=2)

    ax6.plot( x, 'abd_difference_y', data=data, marker='', color='red', linewidth=2)
    ax6.plot( x, 'currentdateheight', data=data, marker='', color='skyblue', linewidth=2)
    ax6.plot( x, 'currentdateweight', data=data, marker='', color='green', linewidth=2)
    
    ax1.legend(loc="upper right")
    ax2.legend(loc="upper right")
    ax3.legend(loc="upper right")
    ax4.legend(loc="upper right")    
    ax5.legend(loc="upper right")  
    ax6.legend(loc="upper right") 
    plt.savefig(path+"/Death_Cases/"+uhid+'.png')

```
